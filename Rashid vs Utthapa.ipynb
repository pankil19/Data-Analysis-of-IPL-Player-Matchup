```python
rashid_df= df[df.bowler == 'Rashid Khan']
```

```python
rashid_df.head()
```

```python
rdf1 = pd.DataFrame(rashid_df.groupby('striker')['runs_off_bat'].sum())
```

```python
rdf2 =  pd.DataFrame(rashid_df.groupby('striker')['ball'].count())
```

```python
rdf1
```

```python
rdf2
```

```python
rdf3= rdf1.merge(rdf2, on = 'striker', how = 'left')
```

```python
rdf3['strike_rate'] = 100*rdf3['runs_off_bat']/rdf3['ball']
```

```python
rdf3 = rdf3[rdf3.ball >= 30].reset_index()
```

```python
rdf3
```

```python
rdf4 =  pd.DataFrame(rashid_df.groupby('player_dismissed')['ball'].count()).reset_index()
```

```python
rdf4
```

```python
plt.figure(figsize= (14,7))
plt.scatter(rdf3.strike_rate, rdf3.runs_off_bat)


for i in range(len(rdf3)):
    plt.text(rdf3['strike_rate'][i], rdf3['runs_off_bat'][i], rdf3['striker'][i])
        
     

plt.title('Batsman against Rashid in IPL (min 30 balls faced)')
plt.xlabel('Strike Rate')
plt.ylabel('Runs Scored')
plt.show()

```
